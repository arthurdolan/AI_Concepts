```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("CO2 Emissions_Canada.csv")
print(df.info())
```

```python
df.plot(kind= 'scatter', x = 'Engine Size(L)', y = 'CO2 Emissions(g/km)')
plt.show
```

```python
X=df[['Engine Size(L)']]
Y=df['CO2 Emissions(g/km)']
X_test, X_train, Y_test, Y_train = train_test_split(X,Y, test_size=0.2, random_state= 42)
```

```python
model = LinearRegression().fit(X_train, Y_train)
print(f"The intercept is: {model.intercept_}")
print(f"The gradient is: {model.coef_}")
```

```python
y_pred = model.predict(X_test)

# Print the predictions
print("Predicted CO2 Emissions:", y_pred)

# Optionally, compare with actual values
print("Actual CO2 Emissions:", Y_test.values)
```
