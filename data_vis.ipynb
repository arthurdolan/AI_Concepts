## Task 3.2: Contour Plots
Generate a contour plot of a function to visualise its variation in two dimensions.

1: Create an array x with 50 points linearly spaced from -π to π <br>
2: Set y equal to x for contour generation<br>
3: Create a meshgrid from x and 'y' <br>
4: Define Z using the function Z = cos(Y) / (1 + X^2) <br>
5: Create a contour plot of Z

```python
x = np.linspace(-np.pi, np.pi, 50)
print(x)
y = x
X,Y = np.meshgrid(x,y)
Z = np.cos(Y)/(1+X**2)
plt.contourf(X, Y, Z, cmap = 'viridis')
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

## Task 3.3: 3D Surface Plot

Create a 3D surface plot using the same meshgrid and values for Z in Task 3.2.

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
surface = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
fig.colorbar(surface,ax=ax, shrink = 0.5, aspect = 10)
ax.set_title("3D Surface Plot of Z = X^2 + Y^2")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()
```
